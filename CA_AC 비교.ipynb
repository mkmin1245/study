```python
"""
for E85A
양극 전장 : 313.2 +0.2,-0.3 (312.15 ± 0.25 )
음극 전장 : 317 +0.3,-0.2 (317.05 ± 0.25)
AC_C(라미산포) : 1.8 ± 0.6
stacking : 0 ± 0.45
"""
```

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
CA = []
for i in range(1000):
  cathode_length = 312.15+0.25/4*np.random.randn(i)
  anode_length = 317.05+0.25/4*np.random.randn(i)
  cathode_top = np.array([0 for k in range(i)])
  cathode_bottom = cathode_top+cathode_length
  anode_top = np.array([0 for k in range(i)])
  anode_bottom = anode_top + anode_length
  lamination = 1.8+0.6/4*np.random.randn(i)
  cathode_top_lami = cathode_top+lamination
  cathode_bottom_lami = cathode_bottom+lamination
  AC = cathode_top_lami-anode_top
  lami_cell = np.array([cathode_top_lami, cathode_bottom_lami, anode_top, anode_bottom])
  lami_cell_alignforstacking = lami_cell-AC-cathode_length/2
  stacking = 0.45/4*np.random.randn(i)
  stacking_cell = lami_cell_alignforstacking+stacking
```

```python
lami_cell_alignforstacking[1]-cathode_length/2
```

```python
for j in range(998):
  CA.append(stacking_cell[0][j]-stacking_cell[2][j+1])
```

```python
CA = np.array(CA)
```

```python
plt.scatter(AC[:998],CA, alpha = 0.1)

plt.show()
```
