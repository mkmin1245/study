```
import numpy as np
from matplotlib import pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```
num = 1000

edge_x = np.random.uniform(-3,0,size=num)
edge_y = np.random.uniform(-3,0,size=num)
```

```
def cal_ab(x1,x2,y1,y2) :
  a = (y2-y1)/(x2-x1)
  b = (x2*y1-x1*y2)/(x2-x1)
  return a,b
```

```
a, b = cal_ab(edge_x,0, 0, edge_y)
```

```
mirror_x = (-2*a*b)/(a*a+1)
mirror_y = 2*b/(a*a+1)
```

```
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np


plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 4)
plt.rcParams['font.size'] = 12

fig, ax = plt.subplots()

x1, x2, mx, y1, y2, my = edge_x, np.zeros(1000), mirror_x, np.zeros(1000), edge_y, mirror_y
ax.set(xlim=[-4, 1], ylim=[-4,1])

ax.add_patch(
     patches.Polygon(
         ((-4,0), (0,0), (0,-4), (-4,-4)),
         edgecolor = 'black',
         facecolor = 'lightgray',
         fill=True,
   ))


for i in range(1000):

  ax.add_patch(
     patches.Polygon(
         ((x1[i], y1[i]), (x2[i], y2[i]), (mx[i], my[i])),
         edgecolor = 'deeppink',
         facecolor = 'lightgray',
         fill=True,
   ))

mirrorpoint = ax.scatter(mx, my, color = "blue", s= 1, alpha = 0.05)
#mirrorpoint = ax.scatter(mirror_x, mirror_y)

plt.show()
```
